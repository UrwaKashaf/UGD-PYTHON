## SERIES

Series is a data structure in python language.Series is a complex data structure.It is used to store collections.
A data structure in pandas that holds an array (collection) of information along with a named index
the names index differentiates this from a simple numpy array.
Formal Definition : one dimensional ndarray with axis label

```python
# make a simple calculator
# + - / *

num1 = int(input("enter any number"))
operator = input("enter any operator ( +, -, /, * )")
num2 = int(input("enter any number"))

if(operator == "+"):
    result = num1 + num2
    print(result)
elif(operator == "-"):
    result = num1 - num2
    print(result)
elif(operator == "*"):
    result = num1 * num2
    print(result)
elif(operator == "/"):
    result = num1 / num2
    print(result)
```
